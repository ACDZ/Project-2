# Outliers Removal and EDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statistics
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.metrics import r2_score

# Set random seed.
np.random.seed(42)


%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
df_train = pd.read_csv("../data/train_dummies_matched.csv")
pd.set_option('display.max_columns', 4000)
pd.set_option('display.max_rows', 4000)
```

```python
# The first thing to look at for cleaning of outliers is to look at sale prices that are way off
```

```python
#It can be seen that the sale price skews to the right.
#As such, some cleaning will be done to remove the outliers.
df_train.hist(column = "SalePrice")
```

```python
# Cleaning will be mainly done by removing values above (mean + 3*std dev)
df_in = df_train[(df_train["SalePrice"] < np.mean(df_train["SalePrice"]) + 3*np.std(df_train["SalePrice"]))]
```

```python
#It has removed 500,000 and 600,000 prices in the skew
df_in.hist(column = "SalePrice")
```

### Important factors
