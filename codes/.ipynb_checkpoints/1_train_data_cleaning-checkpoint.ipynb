#### As there are alot of null in Lot Frontage, it has to be cleaned
#### unfortunately the column is a numerical column and not categorical
#### I made the decision to include the mean because it was common practice
#### using this, I can tell the client that the data is not complete, but by looking at the stats, there are 83.9% of houses with frontage and the mean of lot frontages is 69
#### The decision now is to either fill the blank cells with 0 or to fill it up with the mean

```python
#details = ['Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Fireplace Qu', 'Garage Cond', 'Garage Finish', 'Garage Qual', 'Garage Type', 'BsmtFin Type 1', 'BsmtFin Type 2', 'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', "Mas Vnr Type"]
details = ['Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Fireplace Qu',  "Mas Vnr Type", 'Garage Type']
```

```python
modded_df.isnull().sum().sort_values(ascending = False).head(10)
```

```python
details_ord = ['Kitchen Qual', 'Heating QC', 'Exter Qual', 'Exter Cond', 'Bsmt Qual', 'Bsmt Cond', 'Fireplace Qu', 'Garage Qual', 'Garage Cond', 'Pool QC']
```

```python
for i in details_ord:
    modded_df[i].replace("Ex", 5, inplace = True)
    modded_df[i].replace("Gd", 4, inplace = True)
    modded_df[i].replace("TA", 3, inplace = True)
    modded_df[i].replace("Fa", 2, inplace = True)
    modded_df[i].replace("Po", 1, inplace = True)
    modded_df[i].replace([np.nan], 0, inplace = True)
```

```python
# renaming ordinal values for lot shape
# Reg = 4
# IR1 = 3
# IR2 = 2
# IR3 = 1

lot_shape_dict = {"Reg": 4, "IR1": 3, "IR2": 2, "IR3": 1}
modded_df.replace({"Lot Shape": lot_shape_dict}, inplace = True)
```

```python
# renaming ordinal values for utilities
# AllPub = 4
# NoSewr = 3
# NoSeWa = 2
# ELO = 1

utilities_dict = {"AllPub": 4, "NoSewr": 3, "NoSeWa": 2, "ELO": 1}
modded_df.replace({"Utilities": utilities_dict}, inplace = True)
```

```python
# renaming ordinal values for Land Slope
# Gtl = 3
# Mod = 2
# Sev = 1

land_slope_dict = {"Gtl": 3, "Mod": 2, "Sev": 1}
modded_df.replace({"Land Slope": land_slope_dict}, inplace = True)
```

```python
# renaming ordinal values for Basement Exposure (Bsmt Exposure)
# Gd = 4
# Av = 3
# Mn = 2
# No = 1

bsmt_exp_dict = {"Gd": 4, "Av": 3, "Mn": 2, "No": 1, np.nan: 0}
modded_df.replace({"Bsmt Exposure": bsmt_exp_dict}, inplace = True)
```

```python
# renaming ordinal values for Basement Finish Type 1 (BsmtFin Type 1)
# GLQ = 6
# ALQ = 5
# BLQ = 4
# Rec = 3
# LwQ = 2
# Unf = 1

bsmt_f1_dict = {"GLQ": 6, "ALQ": 5, "BLQ": 4, "Rec": 3, "LwQ": 2, "Unf": 1, np.nan: 0}
modded_df.replace({"BsmtFin Type 1": bsmt_f1_dict}, inplace = True)
```

```python
# renaming ordinal values for Basement Finish Type 2 (BsmtFin Type 2)
# GLQ = 6
# ALQ = 5
# BLQ = 4
# Rec = 3
# LwQ = 2
# Unf = 1

bsmt_f2_dict = {"GLQ": 6, "ALQ": 5, "BLQ": 4, "Rec": 3, "LwQ": 2, "Unf": 1, np.nan: 0}
modded_df.replace({"BsmtFin Type 2": bsmt_f2_dict}, inplace = True)
```

```python
# renaming ordinal values for Electrical System 
#(Electrical)
# SBrkr = 5
# FuseA = 4
# FuseF = 3
# FuseP = 2
# Mix = 1

electrical_dict = {"SBrkr": 5, "FuseA": 4, "FuseF": 3, "FuseP": 2, "Mix": 1}
modded_df.replace({"Electrical": electrical_dict}, inplace = True)
```

```python
# renaming ordinal values for Home Functionality
#(Functional)
# Typ = 8
# Min1 = 7
# Min2 = 6
# Mod = 5
# Maj1 = 4
# Maj2 = 3
# Sev = 2
# Sal = 1

functional_dict = {"Typ": 8, "Min1": 7, "Min2": 6, "Mod": 5, "Maj1": 4, "Maj2": 3, "Sev": 2, "Sal": 1}
modded_df.replace({"Functional": functional_dict}, inplace = True)
```

```python
# renaming ordinal values for Garage Finish
#(Garage Finish)
# Fin = 3
# RFn = 2
# Unf = 1

garage_f_dict = {"Fin": 3, "RFn": 2, "Unf": 1, np.nan: 0}
modded_df.replace({"Garage Finish": garage_f_dict}, inplace = True)
```

```python
# renaming ordinal values for Paved Driveway
#(Paved Drive)
# Y = 3 = Paved
# P = 2 = Partial
# N = 1 = Dirt/Gravel

paved_d_dict = {"Y": 3, "P": 2, "N": 1}
modded_df.replace({"Paved Drive": paved_d_dict}, inplace = True)
```

```python
# renaming ordinal values for Fence Quality
#(Fence)
# GdPrv = 4 = Good Privacy
# MnPrv = 3 = Minimum Privacy
# GdWo = 2 = Good Wood
# MnWw = 1 = Minimum Wood/Wire

fence_dict = {"GdPrv": 4, "MnPrv": 3, "GdWo": 2, "MnWw": 1}
modded_df.replace({"Fence": fence_dict}, inplace = True)
```
